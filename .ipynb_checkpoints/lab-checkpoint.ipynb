```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import sklearn
%matplotlib inline
```

```python
# PNGデータの取り込み
data = mpimg.imread('./lena_gray.png')
```

```python
# 訓練用点群の取り出し
coor = []
val = []
for i in range(5000):
    x,y = np.random.randint(0,256),np.random.randint(0,256)
    coor.append([x,y])
    val.append(data[255-y][x])
X = np.array(coor)
Y = np.array(val)
print(X.shape,Y.shape)
print(Y.min(),Y.max())
```

```python
# 元データと訓練用点群の作図
fig = plt.figure(figsize=(10,5))
ax2 = fig.add_subplot(1,2,1)
ax2 = plt.imshow(data)
ax=fig.add_subplot(1,2,2)
ax.scatter(X[:,0],X[:,1],c=Y[:],marker='.')
plt.show()
```

```python
# Decision tree Regressionで回帰
from sklearn import tree
clf = tree.DecisionTreeRegressor(max_depth=10).fit(X,Y)
out = []
for _x in range(256):
    for _y in range(256):
        out.append([_x,_y,clf.predict([[_x,_y]])[0]])
outnp = np.array(out)
```

```python
fig2 = plt.figure(figsize=(10,5))
ax_or = fig2.add_subplot(1,2,1)
ax_or = plt.imshow(data)
ax_pred = fig2.add_subplot(1,2,2)
ax_pred.scatter(outnp[:,0],outnp[:,1],c=outnp[:,2],marker='.')
plt.show()
```
